# Process Confusion Matrices

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

fp_checkpoint = os.path.join("../checkpoints")
fp_cm_folder = os.path.join(fp_checkpoint, "confusion_matrices")

# 
```

```python
def generate_confusion_matrix(filename, vmin=0, vmax=100, font_scale=0.7, figsize=(2, 2)):
    fp = os.path.join(fp_cm_folder, filename)
    df = pd.read_csv(fp, index_col=0, keep_default_na=False)
    # display(df)
    fig, ax = plt.subplots(1, 1, figsize=figsize, dpi=300)
    sn.set(font_scale=font_scale) # for label size
    sn.heatmap(df, ax=ax, cmap="Blues", annot=True, fmt='g', vmin=vmin, vmax=vmax, cbar=False)
    condition = filename.split("_")[0][:-1]
    ax.set_title(condition, fontsize=14)
    label_size = 8
    ax.set_xlabel('Prediction', fontsize=label_size)
    ax.set_ylabel('Groundtruth', fontsize=label_size)
    plt.savefig(f"{condition}_confusion_matrices.jpg", bbox_inches="tight")
    plt.tight_layout()
for condition in ["AMDC", "DMC", "DRC", "GSC"]:
    generate_confusion_matrix(
        filename=f"{condition}_confusion_matrices_test.csv", vmax=157,
        # font_scale=0.7 if (condition in ["DMC", "DRC"]) else 0.5,
        # figsize= (3, 3) if condition == "DRC" else (2,2)
    )
```
